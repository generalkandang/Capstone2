```python
#load our model

Pkl_Filename = "Lol_model.pkl" 
with open(Pkl_Filename, 'rb') as file:  
    model = pickle.load(file)
```

```python
key_features = ['kills_pg', 'towers_pg', 'assists_pg', 'firsttower_pg', 'earnedgold_pg', 'inhibitors_pg', 'Win_Percentage']
```

```python
X = df.loc[df.index.get_level_values(0) != "2020", key_features]
y = df.loc[df.index.get_level_values(0) != "2020", 'Championship'].values == 1
```

```python
len(X), len(y)
```

```python
model.fit(X, y)
```

```python
#see accuracy on data excluding 2020
from sklearn.metrics import accuracy_score
print(accuracy_score(model.predict(X), y))
```
